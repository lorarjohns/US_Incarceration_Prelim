# Hypothesis 3: The rate of exoneration for murder in a state is correlated to the rate of incarceration for murder in that state

## Step 1: Set up null and alternative hypotheses

#### Ho: The exoneration for murder rate is not correlated with the incarceration for murder rate

#### Ha: The exoneration for murder rate is correlated with the incarceration for murder rate


## Step 2: Choose a significance level

```python
alpha = 0.05
```

## Step 3: Calculate the test statistic

### Step 3.1: Import Necessary Libraries

```python
# Import Pandas
import pandas as pd

# Import Stats from Scipy
from scipy import stats

# Import Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Import numpy
import numpy as np
```

### Step 3.2: Load, clean and explore datasets

```python
# Load in and preview first dataset, 'NIJ_exonerations.xlsx'
# This data includes case by case exoneration information in the United States from 1989 to 2019

df1 = pd.read_excel('NIJ_exonerations.xlsx')
df1.head()
```

```python
# Check Columns to see where date exonerated is

df1.columns
```

```python
df1['Exonerated'].head()
```

```python
# Check all the unique values in 'Worst Crime Display' to ensure they're are no 
# duplicates for murder or differing values

df1.groupby('Worst Crime Display')['ID'].nunique() # murder - 934
```

```python
# Create a dataframe containing only murder exonerations between 2001 and 2016 (Our other dataset only 
# has values between 2001 and 2016)

murder_ex_df = df1[(df1['Worst Crime Display'] == 'Murder') & (df1['Exonerated'] > 2000) & (df1['Exonerated'] < 2017)]
murder_ex_df.head()
```

```python
# Check Count Number of Murder exonerations there are in the dataset

len(murder_ex_df)
```

```python
# Count the number of murder exonerations per State

murder_ex_df = murder_ex_df.groupby('State', as_index=False).agg({"Worst Crime Display": "count"})
murder_ex_df = murder_ex_df.rename(columns={'Worst Crime Display': 'murder_exoneration_count'})
murder_ex_df['State'] = murder_ex_df['State'].str.upper() 
murder_ex_df
```

```python
# Add Hawaii, Maine, New Hampshire, North Dakota, South Dakota, Maine, Delaware
listOfSeries = [pd.Series(['HAWAII', 0], index=murder_ex_df.columns ), 
                pd.Series(['NEW HAMPSHIRE', 0], index=murder_ex_df.columns), 
                pd.Series(['NORTH DAKOTA', 0], index=murder_ex_df.columns),
                pd.Series(['SOUTH DAKOTA', 0], index=murder_ex_df.columns),
                pd.Series(['MAINE', 0], index=murder_ex_df.columns),
                pd.Series(['DELAWARE', 0], index=murder_ex_df.columns)]


murder_ex_df = murder_ex_df.append(listOfSeries , ignore_index=True)
```

```python
# Import second dataset, 'crime_and_incarceration_by_state.csv'
# This dataset includes information on number of incarcerations in each state for different types of crimes
# as well as state population information

crime_and_incarceration_df = pd.read_csv('crime_and_incarceration_by_state.csv')
crime_and_incarceration_df.head()
```

```python
# Select subset with only needed items, 'jurisdiction', 'murder_manslaughter', and 'state_population'

crime_and_incarceration_df = crime_and_incarceration_df[['jurisdiction','murder_manslaughter', 'state_population']]
crime_and_incarceration_df.head()
```

```python
# Group crime_and_incarceration_df by 'jurisdiction', andfor each jurisdiction, 
# show the sum of all the 'murder_manslaughter' incarceration between 2001 and 2016,
# also show the average 'state_population' between 2001 and 2016

crime_and_incarceration_df=crime_and_incarceration_df.groupby('jurisdiction').agg({"murder_manslaughter": "sum", "state_population": "mean"})
crime_and_incarceration_df = crime_and_incarceration_df.rename(columns={'murder_manslaughter': 'murder_manslaughter_count'})
crime_and_incarceration_df.head()
```

```python
len(crime_and_incarceration_df)
```

```python
# Merge crime_and_incarceration_df and murder_ex_df

temp_df = pd.merge(crime_and_incarceration_df, murder_ex_df, left_on='jurisdiction', right_on='State')
temp_df.head()
```

```python
# Create final dataset to be analyzed

final_df = temp_df[['State', 'murder_exoneration_count', 'murder_manslaughter_count', 'state_population']]
final_df['exoneration_rate'] = final_df['murder_exoneration_count']/final_df['state_population']
final_df['incarceration_rate'] = final_df['murder_manslaughter_count']/final_df['state_population']
final_df.head()
```

```python
final_df # Please not, there are no values for Federal, District of Columbia or Puerto Rico in this dataset
```

Since we are looking to see how our two columns correlate with each other, 'exoneration_rate' and 'incarceration_rate' we will not use a z-test or a t-test to determine significance.  Instead we will use scipy.stats.pearsonr(a,b) to determine Pearson’s correlation coefficient and the 2-tailed p-value. Therefore we do not need a test statistic.


## Step 4: Determine the Critical or P-value

```python
from scipy import stats
pearsons_r, two_tail_p_value = stats.pearsonr(final_df.exoneration_rate, final_df.incarceration_rate)
```

```python
print("Pearson's R: {}, Two-tailed p-value: {}".format(pearsons_r, two_tail_p_value))
```

```python
plt.scatter(final_df.exoneration_rate, final_df.incarceration_rate)
plt.ylim(0,.002)
plt.xlim(-.000001, .000009)
plt.xlabel('Exoneration Rate (Exoneration Count/Average State Population)')
plt.ylabel('Incarceration Rate (Incarceration Count/Average State Population)')
plt.title('Murder Exoneration Rate Vs. Murder Incarceation Rate (2001-2016)')
```

## Step 5: Reject or Fail to Reject the Null Hypothesis

Since our alpha value is 0.05 and our calculated p-value is 0.021343638483385798, we can reject our null hypothesis. The exoneration rate for a state is positively correlated by 0.3248 to the incarceration rate for a state.

States that have a higher rate of exoneration for murder are seeing higher rates of incarceration for murder.  This would make it appear that crime, specifically murder is more prevalent in states who's criminal justice systems have failed it's people.
